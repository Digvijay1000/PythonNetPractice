### Data Types

###### why we need data types ?
1. **Memory management** : 
2. **Data Integritity and validation** : Data types help ensure that the operations performed on the data are meaningful. For example, you can add two integers together, but adding a string to an integer usually doesn't make sense and would raise an error. This prevents many types of logical errors.
3. **Operations and Methods** : Different data types support different operations and methods. For instance, strings support operations like concatenation and methods like .upper(), while lists support operations like indexing and methods like .append(). 
4. **Code Clarity and Maintainability** : Explicit data types make the code more readable and easier to understand. When you see a variable with a specific type, you get an idea of what kind of data it holds and what operations are appropriate for it. This clarity is especially important in large codebases and when working in teams.
5. **Performance Optimization** : Some operations are more efficient with certain data types. For example, operations on integers are generally faster than operations on floating-point numbers due to the simpler arithmetic. Understanding the data types can help optimize the performance of the program.
6. **Interfacing with Other Systems** : When interacting with databases, APIs, or other systems, data often needs to be in a specific format. Data types help in ensuring that the data being sent or received matches the expected format, facilitating smooth communication between different systems.
7. **Error Prevention** : Using the correct data type can prevent a lot of runtime errors. For example, trying to access an index in a list that does not exist will raise an error, while iterating over a dictionary requires handling keys properly. Knowing the data type helps in writing error-free code.

**PEMDAS** is an acronym used to remember the order of operations in mathematics. It stands for:

Parentheses: Perform operations inside parentheses first.

Exponents: Evaluate exponents (powers and roots, such as squares and square roots).

Multiplication and Division: Perform multiplication and division from left to right.

Addition and Subtraction: Perform addition and subtraction from left to right.

Here's how PEMDAS is applied step-by-step:

Parentheses: Simplify expressions inside parentheses (or brackets) first.

Example: 

```python
print(3*3 +3/3 -3) 
# 1. 3*3 + 3/3 - 3
# 2. 9 + 3/3 - 3
# 3. 9 + 1 - 3
# 4. 10 - 3
# 5. 7 
```

###### Interactive Coding Exercise -  BMI Calculator

Write a program that calculates the Body Mass Index (BMI) from a user's weight and height.



```python
# 1st input: enter height in meters e.g: 1.65
height = input("Enter Height in meter")
# 2nd input: enter weight in kilograms e.g: 72
weight = input("Enter weight in kg")
# 🚨 Don't change the code above 👆

# Write your code below this line 👇
height_as_float = float(height)
weight_as_int = int(weight)
bmi = weight_as_int / height_as_float ** 2
# 0r 
bmi = weight_as_int / (height_as_float * height_as_float)

bmi_as_int = int(bmi)
print(bmi_as_int)
```

### Day 2 Project - Tip Calculator

```python
#If the bill was $150.00, split between 5 people, with 12% tip. 

#Each person should pay (150.00 / 5) * 1.12 = 33.6
#Format the result to 2 decimal places = 33.60
```

```python
print("Welcome to the tip calculator!")
bill = float(input("What was the total bill? : $"))
tip = int(input("How much tip would you like to give? 10%, 12%, or 15%? : "))
people = int(input("How many people to split the bill? : "))

tip_as_percentage = tip / 100
total_tip_amount = bill * tip_as_percentage

total_bill = bill + total_tip_amount
bill_per_person = total_bill / people
print(f"Each person should pay : ${round(bill_per_person,2)}")
```
