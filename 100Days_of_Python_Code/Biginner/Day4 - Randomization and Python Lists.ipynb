Extend the list by appending all the items from the iterable.

###### list.insert(i,x) :
Insert an item at a given position. The first argument is the index of the element before which to insert.

```python
fruits.insert(2, "Badam")
print(fruits)
```

###### list.remove(x) :
Remove the first item from the list whose value is equal to x. It raises a ValueError if there is no such item.

```python
fruits.remove("Badam")
print(fruits)
```

###### list.pop( i ) :

Remove the item at the given position in the list, and return it. If no index is specified, a.pop() removes and returns the last item in the list. It raises an IndexError if the list is empty or the index is outside the list range.

```python
fruits.pop(2)
print(fruits)
```

###### list.sort(*, key=None, reverse=False)

Sort the items of the list in place (the arguments can be used for sort customization, see sorted() for their explanation).

The sort method does not return a new list. Instead, it modifies the original list in place.

If a key function is provided, the sort method uses the values returned by the key function to sort the list.

If reverse is True, the list is sorted in descending order.

```python
my_list = [3, 1, 4, 1, 5, 9, 2, 6, 5]
my_list.sort()
print(my_list)  # Output: [1, 1, 2, 3, 4, 5, 5, 6, 9]
```

```python
my_list = [3, 1, 4, 1, 5, 9, 2, 6, 5]
my_list.sort(reverse=True)
print(my_list)  # Output: [9, 6, 5, 5, 4, 3, 2, 1, 1]
```

```python
my_list = ["apple", "banana", "cherry", "date"]
my_list.sort(key=len) # sort them by their lengths
print(my_list)  # Output: ['date', 'apple', 'banana', 'cherry']
```

```python
my_list = ["apple", "banana", "cherry", "date"]
my_list.sort(key=len, reverse=True)
print(my_list)  # Output: ['banana', 'cherry', 'apple', 'date']
```

#### Shallo copy and Deep Copy



```python
import copy

original_list = [1, 2, [3, 4]]
shallow_copied_list = copy.copy(original_list)

shallow_copied_list[2][0] = 'changed'
print(original_list)        # Output: [1, 2, ['changed', 4]]
print(shallow_copied_list)  # Output: [1, 2, ['changed', 4]]
```

```python
import copy

original_list = [1, 2, [3, 4]]
deep_copied_list = copy.deepcopy(original_list)

deep_copied_list[2][0] = 'changed'
print(original_list)        # Output: [1, 2, [3, 4]]
print(deep_copied_list)     # Output: [1, 2, ['changed', 4]]
```
